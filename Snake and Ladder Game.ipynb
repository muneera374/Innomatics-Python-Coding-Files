```python
import random
token1=0
token2=0
count1=0
count2=0
Ladders={1:38,4:14,9:31,28:84,80:99,72:91}
Snakes={17:7,54:34,62:19,64:60,87:36,92:73,95:75,98:79}
while token1<100 or token2<100:
    input("player1 press Enter to roll dice")
    dice=random.randint(1,6)
    print("{},player 1 rolled the dice".format(dice))
    if dice==1:
        count1=count1+1
        if count1==1:
            print("Yay Player1 started the game")
    if token1>=0 and token1<=100 and count1>=1:
        token1=token1+dice
        if token1 in Ladders:
            print("Yay There is a ladder at {}".format(token1))
            token1=Ladders[token1]
        if token1 in Snakes:
            print("Oh no There is a Snake at {}".format(token1))
            token1=Snakes[token1]
        while dice==6:
            if token1>100:
                token1=token1-dice
            else:
                print("Player1 is at position {}".format(token1))
                input("player1 press Enter to roll dice")
                dice=random.randint(1,6)
                print("{},player 1 rolled the dice".format(dice))
                token1=token1+dice
        if token1>100:
            token1=token1-dice
            print("Player1 is at position {}".format(token1))
        else:
            print("Player1 is at position {}".format(token1))
    if token1==100:
        print("Congratulations Player1 won the game")
        break
    print("\n")
    input("player2 press Enter to roll dice")
    dice=random.randint(1,6)
    print("{},player 2 rolled the dice".format(dice))
    if dice==1:
        count2=count2+1
        if count2==1:
            print("Yay Player2 started the game")
    if token2>=0 and token2<=100 and count2>=1:
        token2=token2+dice
        if token2 in Ladders:
            print("Yay There is a ladder at {}".format(token2))
            token2=Ladders[token2]
        if token2 in Snakes:
            print("Oh no There is a Snake at {}".format(token2))
            token2=Snakes[token2]
        while dice==6:
            if token2>100:
                token2=token2-dice
            else:
                print("Player2 is at position {}".format(token2))
                input("player2 press Enter to roll dice")
                dice=random.randint(1,6)
                print("{},player 2 rolled the dice".format(dice))
                token2=token2+dice
        if token2>100:
            token2=token2-dice
            print("Player2 is at position {}".format(token2))
        else:
            print("Player2 is at position {}".format(token2))
    if token2==100:
        print("Congratulations Player 2 won the game")
        break
    print("\n")
```
